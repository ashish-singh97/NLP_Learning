### Stemming
Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer, RegexpStemmer, ISRIStemmer, RSLPStemmer, WordNetLemmatizer
```

```python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
# Porter Stemmer
porter_stemmer=PorterStemmer()
for word in words:
    print(word,"--->",porter_stemmer.stem(word))
```

```python
porter_stemmer.stem('congratulations')
```

```python
porter_stemmer.stem("sitting")
```

### Snowball Stemmer
It is a stemming algorithm which is also known as the Porter2 stemming algorithm as it is a better version of the Porter Stemmer since some issues of it were fixed in this stemmer.

```python
snowballsstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word,"--->",snowballsstemmer.stem(word))
```

```python
#WordNet Lemmatizer
wordnet_lemmatizer=WordNetLemmatizer()
for word in words:
    print(word,"--->",wordnet_lemmatizer.lemmatize(word))
```

```python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''
print(wordnet_lemmatizer.lemmatize("went",pos='v'))
print(wordnet_lemmatizer.lemmatize("going",pos='v'))
print(wordnet_lemmatizer.lemmatize("gone",pos='v'))
print(wordnet_lemmatizer.lemmatize("better",pos='a'))
```

### RegexpStemmer class
NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

| Class                   | One-line explanation                                                                                       |
| ----------------------- | ---------------------------------------------------------------------------------------------------------- |
| **`PorterStemmer`**     | The most widely used English stemmer that removes common suffixes to produce a root-like word.             |
| **`LancasterStemmer`**  | A more aggressive English stemmer that often produces shorter stems than Porter.                           |
| **`SnowballStemmer`**   | An improved, multilingual stemmer that supports multiple languages and is generally preferred over Porter. |
| **`RegexpStemmer`**     | A customizable stemmer that removes word endings based on user-defined regular expression patterns.        |
| **`ISRIStemmer`**       | A stemmer specifically designed for Arabic text using the ISRI stemming algorithm.                         |
| **`RSLPStemmer`**       | A rule-based stemmer for Portuguese that removes suffixes according to RSLP linguistic rules.              |
| **`WordNetLemmatizer`** | A lemmatizer that uses the WordNet dictionary to return the actual base (dictionary) form of a word.       |


#### Example:
| Input     | Porter   | Lancaster | Snowball | Lemmatizer                      |
| --------- | -------- | --------- | -------- | ------------------------------- |
| `running` | `run`    | `run`     | `run`    | `run` *(with POS=`verb`)*       |
| `studies` | `studi`  | `study`   | `studi`  | `study`                         |
| `better`  | `better` | `bet`     | `better` | `good` *(with POS=`adjective`)* |
| `cars`    | `car`    | `car`     | `car`    | `car`                           |
