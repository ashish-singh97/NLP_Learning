# TF-IDF (Term Frequency – Inverse Document Frequency)

## Definition

**TF-IDF** is a feature extraction technique that converts text into numerical vectors by measuring the importance of a word in a document relative to an entire collection of documents (corpus).

Unlike **Bag of Words**, TF-IDF reduces the importance of common words and increases the importance of rare but meaningful words.

---

# Formula

## 1. Term Frequency (TF)

**Definition:**
Measures how frequently a word appears in a document.

### Formula

\[
TF(t,d)=\frac{\text{Number of times term }t\text{ appears in document }d}{\text{Total number of terms in document }d}
\]

### Example

Document:

```
I love NLP and I love Python
```

Word Count:

| Word | Count |
|------|------:|
| I | 2 |
| love | 2 |
| NLP | 1 |
| and | 1 |
| Python | 1 |

Total words = **7**

TF of **love**

```
TF = 2 / 7 = 0.286
```

---

# 2. Inverse Document Frequency (IDF)

**Definition:**
Measures how unique or rare a word is across all documents.

### Formula

\[
IDF(t)=\log\left(\frac{N}{DF(t)}\right)
\]

where

- **N** = Total number of documents
- **DF(t)** = Number of documents containing the term

---

### Example

Corpus:

```
Doc1: I love NLP
Doc2: I love Machine Learning
Doc3: NLP is amazing
```

Total documents

```
N = 3
```

Word **love** appears in

```
Doc1
Doc2
```

Document Frequency

```
DF = 2
```

IDF

```
log(3 / 2)
```

Word **Machine** appears only once.

```
DF = 1
```

IDF

```
log(3 / 1)
```

Since **Machine** appears in fewer documents, it receives a higher IDF score than **love**.

---

# 3. TF-IDF Score

**Definition:**

TF-IDF is the product of TF and IDF.

### Formula

\[
TFIDF = TF \times IDF
\]

Words that occur frequently in a document but rarely across the corpus receive the highest scores.

---

# Example

Documents

```
Doc1: I love NLP

Doc2: I love Machine Learning

Doc3: NLP is amazing
```

### Bag of Words

| Word | Doc1 | Doc2 | Doc3 |
|------|-----:|-----:|-----:|
| I | 1 | 1 | 0 |
| love | 1 | 1 | 0 |
| NLP | 1 | 0 | 1 |
| Machine | 0 | 1 | 0 |
| Learning | 0 | 1 | 0 |
| is | 0 | 0 | 1 |
| amazing | 0 | 0 | 1 |

Every word contributes equally.

---

### TF-IDF

Common words such as

```
I
love
is
```

receive **lower scores**.

Important words such as

```
Machine
Learning
amazing
```

receive **higher scores** because they are more unique.

---

# Why TF-IDF?

Suppose we have two documents.

```
Doc1: The movie is good.

Doc2: The food is good.
```

Words like

```
the
is
good
```

appear frequently in many documents.

TF-IDF assigns them lower importance.

Words like

```
movie
food
```

appear in fewer documents and receive higher importance.

---

# Advantages

- Reduces the importance of common words.
- Highlights important keywords.
- Simple and computationally efficient.
- Works well for text classification and information retrieval.

---

# Disadvantages

- Ignores word order.
- Does not capture semantic meaning.
- Produces sparse vectors.
- Cannot understand synonyms or context.

---

# TF-IDF vs Bag of Words

| Bag of Words | TF-IDF |
|---------------|---------|
| Counts word frequency. | Assigns importance based on frequency and rarity. |
| Common words have high counts. | Common words receive lower weights. |
| Ignores document frequency. | Considers document frequency. |
| Simpler representation. | More informative representation. |


# Interview Definitions

### Term Frequency (TF)

Measures how often a word appears in a document.

---

### Inverse Document Frequency (IDF)

Measures how rare or unique a word is across the entire corpus.

---

### TF-IDF

A feature extraction technique that assigns higher weights to important words and lower weights to commonly occurring words.

---

# Interview Questions

### What is TF-IDF?

TF-IDF is a statistical technique used to measure the importance of a word in a document relative to an entire corpus.

---

### Why is TF-IDF better than Bag of Words?

Because it reduces the weight of common words and emphasizes informative words, leading to better feature representation.

---

### What is the range of TF-IDF values?

Typically:

```
0 ≤ TF-IDF ≤ 1
```

although the exact range depends on the TF and IDF formulas and any normalization applied by the implementation.

---

### Where is TF-IDF used?

- Text Classification
- Spam Detection
- Search Engines
- Information Retrieval
- Document Similarity
- Keyword Extraction
- Recommendation Systems

---

# Key Takeaway

- **TF** measures how frequently a word appears in a document.
- **IDF** measures how unique a word is across all documents.
- **TF-IDF = TF × IDF**.
- Common words receive **lower weights**.
- Rare and informative words receive **higher weights**.
- TF-IDF generally provides better text features than Bag of Words for many traditional machine learning models.

````python
# Scikit-learn Example
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "I love NLP",
    "I love Machine Learning",
    "NLP is amazing"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
# TF-IDF with N-Grams
#Unigrams only
TfidfVectorizer(ngram_range=(1,1))

#Bigrams only
TfidfVectorizer(ngram_range=(2,2))

#Unigrams + Bigrams
TfidfVectorizer(ngram_range=(1,2))
````

````python
import pandas as pd
````

````python
messages = pd.read_csv("smsspamcollection/SMSSpamCollection.txt", sep='\t', names = ['labels','message'])
````

````python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
````

````python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordlemmatize=WordNetLemmatizer()
````

````python
corpus = []
for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [wordlemmatize.lemmatize(word)for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
````

````python

from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
tfidf=TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
````
