```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize, wordpunct_tokenize, regexp_tokenize, blankline_tokenize, TweetTokenizer, MWETokenizer, WordPunctTokenizer, WhitespaceTokenizer, TreebankWordTokenizer, PunktSentenceTokenizer
```

```python
nltk.download('punkt_tab')  #is a pre-trained sentence and word tokenization model used by NLTK.
```

```python
corpus="""Hello Welcome,to My NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
#Sentence Tokenization
sent = sent_tokenize(corpus)
print(sent)
for sent in sent:
    print(sent)
print(type(sent))
```

```python
#Word Tokenization
words = word_tokenize(corpus)
print(words)
print(type(words))
```

```python
# WordPunct Tokenization
word_punct = wordpunct_tokenize(corpus)
print(word_punct)
```

| Tokenizer                      | One-line explanation                                                                        |
| ------------------------------ | ------------------------------------------------------------------------------------------- |
| **`word_tokenize()`**          | Splits text into words and punctuation using NLTK's recommended tokenizer.                  |
| **`sent_tokenize()`**          | Splits a paragraph into individual sentences.                                               |
| **`wordpunct_tokenize()`**     | Splits words and all punctuation into separate tokens using simple rules.                   |
| **`regexp_tokenize()`**        | Tokenizes text based on a custom regular expression pattern.                                |
| **`blankline_tokenize()`**     | Splits text into blocks separated by one or more blank lines.                               |
| **`TweetTokenizer()`**         | Tokenizes tweets while preserving hashtags, mentions, emojis, and URLs.                     |
| **`MWETokenizer()`**           | Treats predefined multi-word expressions (e.g., "New York") as a single token.              |
| **`WordPunctTokenizer()`**     | Class-based version of `wordpunct_tokenize()` that separates words and punctuation.         |
| **`WhitespaceTokenizer()`**    | Splits text only on whitespace, leaving punctuation attached to words.                      |
| **`TreebankWordTokenizer()`**  | Penn Treebank-style tokenizer that handles contractions and punctuation accurately.         |
| **`PunktSentenceTokenizer()`** | A trainable sentence tokenizer that identifies sentence boundaries using statistical rules. |
