# Word Embedding using Keras Embedding Layer

## Definition

The **Embedding Layer** in Keras is a trainable neural network layer that converts words (represented as integers) into **dense, low-dimensional vectors**. Unlike pre-trained embeddings (such as Word2Vec or GloVe), the embedding vectors are **learned automatically during model training**.

---

# Why use an Embedding Layer?

Traditional techniques such as **Bag of Words** and **TF-IDF** produce sparse vectors and do not capture semantic meaning.

The Embedding Layer learns dense representations where similar words have similar vectors.

Example:

```
I love NLP

↓

Integer Encoding

↓

[1, 5, 10]

↓

Embedding Layer

↓

[[0.21, -0.43, ...],
 [0.87, 0.15, ...],
 [-0.34, 0.91, ...]]
```

---

# How Embedding Layer Works

The Embedding Layer acts like a lookup table.

Suppose we have a vocabulary of five words:

| Word | Integer |
|-------|---------|
| I | 1 |
| love | 2 |
| NLP | 3 |
| Machine | 4 |
| Learning | 5 |

Initially, the embedding matrix contains random values.

```
Embedding Matrix

Word ID    Embedding

1 → [0.12, 0.43, -0.82]

2 → [-0.45, 0.81, 0.21]

3 → [0.67, -0.15, 0.92]

4 → [0.38, 0.54, -0.18]

5 → [-0.72, 0.16, 0.51]
```

During training, these vectors are updated using backpropagation.

# Advantages

- Learns embeddings automatically.
- Dense representation.
- Reduces dimensionality.
- Improves neural network performance.
- Suitable for deep learning models.

---

# Disadvantages

- Requires labeled training data.
- Needs sufficient training examples.
- Learned embeddings are task-specific.
- Not reusable across different tasks unless saved.

---

# Embedding Layer vs Word2Vec

| Embedding Layer | Word2Vec |
|-----------------|----------|
| Learns during model training | Pre-trained or separately trained |
| Task-specific embeddings | General-purpose embeddings |
| Integrated into neural networks | Independent embedding model |
| Requires labeled data | Can be trained on unlabeled text |

---

# Embedding Layer vs TF-IDF

| TF-IDF | Embedding Layer |
|---------|-----------------|
| Sparse vectors | Dense vectors |
| Frequency-based | Learned representations |
| No semantic meaning | Captures semantic similarity |
| High dimensional | Low dimensional |

# Interview Definitions

### Embedding Layer

A trainable neural network layer that maps integer-encoded words into dense vector representations.

---

### Embedding Matrix

A trainable matrix where each row corresponds to the embedding of one word in the vocabulary.

---

### Embedding Dimension

The number of numerical features used to represent each word (e.g., 50, 100, 200, 300).

---

# Interview Questions

### What is an Embedding Layer?

An Embedding Layer is a trainable lookup table that converts integer-encoded words into dense vector representations.

---

### Why use an Embedding Layer?

Because it learns semantic word representations automatically during training and provides better features for deep learning models.

---

### Is the Embedding Layer pre-trained?

No. By default, embeddings are initialized randomly and learned during model training. However, it can also be initialized with pre-trained embeddings such as Word2Vec or GloVe.

---

### What is the input to an Embedding Layer?

A sequence of integer-encoded words.

---

### What is the output?

A dense vector for every input word.

---

# Key Takeaway

- The Embedding Layer is a **trainable lookup table**.
- It converts integer-encoded words into **dense vectors**.
- Embeddings are learned automatically through **backpropagation**.
- It is widely used with **LSTM**, **GRU**, **CNN**, and **Transformer** models.
- It is one of the most common methods for representing text in deep learning.

````python
# ===========================================
# Word Embedding using Keras Embedding Layer
# ===========================================

# Import Libraries
import numpy as np
import tensorflow as tf

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

print("TensorFlow Version:", tf.__version__)
````

````python
# Sample Corpus

sentences = [
    "I love NLP",
    "I love Machine Learning",
    "Deep Learning is amazing",
    "NLP is fun",
    "Machine Learning is awesome"
]
````

````python
# Tokenization

tokenizer = Tokenizer()

tokenizer.fit_on_texts(sentences)

word_index = tokenizer.word_index

print(word_index)
````

````python
# Convert text into integer sequences

sequences = tokenizer.texts_to_sequences(sentences)

print(sequences)
````

````python
#Padding
max_length = max([len(sentence) for sentence in sequences])
X = pad_sequences(
    sequences,
    maxlen = max_length,
    padding = 'post'
)
print(X)
````

````python
# Vocabulary Size

vocab_size = len(word_index) + 1

print("Vocabulary Size:", vocab_size)
````

````python
# Create Embedding Layer

embedding_dim = 8

model = Sequential()

model.add(
    Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim
    )
)

model.summary()
````

````python
# Generate Embeddings

embeddings = model.predict(X)

print(embeddings.shape)
````

Meaning

5 Sentences
4 Words after Padding
8-Dimensional Embedding

````python
# Embedding of First Sentence

print(embeddings[0])
````

````python
# Embedding of the word "NLP"

nlp_id = tokenizer.word_index["nlp"]

embedding_weights = model.layers[0].get_weights()[0]

print("Word ID:", nlp_id)

print("Embedding Vector:")

print(embedding_weights[nlp_id])
````

````python
# Show embedding of every word

for word, index in tokenizer.word_index.items():
    print(f"{word:12} --> {embedding_weights[index]}")
````
