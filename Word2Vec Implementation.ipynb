# Word2Vec (Word Embeddings)

## Definition

**Word2Vec** is a neural network-based technique that converts words into dense numerical vectors (embeddings) while preserving their semantic and contextual relationships.

Unlike **Bag of Words** and **TF-IDF**, Word2Vec captures the **meaning** of words. Words with similar meanings have similar vector representations.

---

# Why Word2Vec?

Suppose we have the following words:

```
King
Queen
Man
Woman
```

Bag of Words and TF-IDF treat them as completely different words.

Word2Vec learns that:

```
King ≈ Queen

Man ≈ Woman
```

because they often appear in similar contexts.

---

# Basic Idea

Instead of counting words, Word2Vec learns vector representations by looking at neighboring words in a sentence.

Example:

```
The cat sits on the mat.
```

Word2Vec learns:

```
cat → animal

mat → object

sits → action
```

from the surrounding context.

---

# Types of Word2Vec

## 1. CBOW (Continuous Bag of Words)

### Definition

Predicts the **target word** using its surrounding context words.

### Example

Sentence:

```
I love learning NLP
```

Training Example

```
Input (Context Words):

I
love
NLP

↓

Output (Target Word):

learning
```

CBOW uses neighboring words to predict the missing word.

### Advantages

- Faster training.
- Performs well for frequent words.
- Suitable for large datasets.

---

## 2. Skip-Gram

### Definition

Predicts the **surrounding context words** using the target word.

### Example

Sentence:

```
I love learning NLP
```

Training Example

```
Input (Target Word):

learning

↓

Output (Context Words):

I
love
NLP
```

Skip-Gram predicts neighboring words from the center word.

### Advantages

- Better for rare words.
- Produces higher-quality embeddings.
- Works well on smaller datasets.

---

# Word2Vec Architecture

```
Sentence

↓

Tokenization

CBOW / Skip-Gram Neural Network

↓

Dense Word Embeddings

```

# Example

Sentence

```

```

Word Embeddings

| Word | Vector |
|------|------------------------------------|

| love | [-0.18, 0.74, -0.31, ...] |
| NLP | [0.88, 0.15, -0.27, ...] |

Each word is represented by a dense numerical vector.

Typical vector sizes:

```
50

100

200

300
```

# Word Similarity

Example:

```

```

Similarity:

```
King ↔ Queen ✔

Prince ↔ Princess ✔

King ↔ Apple ✘
```

Semantically similar words have vectors that are close in the embedding space.

Word2Vec captures semantic relationships through vector arithmetic.

---

# Advantages

- Captures semantic meaning.
- Learns contextual relationships.
- Dense vector representation.
- Handles synonyms well.
- Lower dimensional than Bag of Words.
- Better feature representation for deep learning.

---

# Disadvantages

- Requires a large corpus.
- Cannot represent unseen words (Out-of-Vocabulary words).
- One fixed vector per word regardless of context.
- Training can be computationally expensive.

---

# Word2Vec vs Bag of Words

| Bag of Words | Word2Vec |
|---------------|----------|
| Counts word frequency. | Learns word meaning. |
| Sparse vectors. | Dense vectors. |
| Ignores context. | Uses surrounding context. |
| High-dimensional representation. | Low-dimensional embeddings. |

---

# Word2Vec vs TF-IDF

| TF-IDF | Word2Vec |
|---------|----------|
| Measures word importance. | Learns semantic meaning. |
| Sparse vectors. | Dense vectors. |
| Ignores context. | Learns contextual relationships. |
| High-dimensional. | Compact embeddings. |

# Key Takeaway

- Word2Vec learns **word meanings** instead of counting word frequencies.
- It produces **dense vector representations**.
- It uses **CBOW** or **Skip-Gram** architectures.
- Similar words have similar embeddings.
- Unlike BoW and TF-IDF, it captures **semantic relationships** between words.
- Each word has a **fixed embedding**, which is why newer models like **BERT** were developed to generate **contextual embeddings**.

````python
import gensim
````

````python
from gensim.models import word2vec

from gensim.models import Word2Vec

sentences = [
    ["i", "love", "nlp"],
    ["i", "love", "machine", "learning"],
    ["nlp", "is", "amazing"],
    ["machine", "learning", "is", "fun"]
]

model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)
````

````python
#get embedding of a word
vector = model.wv["nlp"]
print(vector)

print(vector)
````

````python
#find most similar word
model.wv.most_similar("nlp")
````

````python
#similarity b/w two words
model.wv.similarity("nlp", "learning")
````

````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

words = model.wv.index_to_key
vectors = model.wv[words]

pca = PCA(n_components=2)
result = pca.fit_transform(vectors)

plt.figure(figsize=(8,6))

for i, word in enumerate(words):
    plt.scatter(result[i,0], result[i,1])
    plt.text(result[i,0], result[i,1], word)

plt.show()
````
